```python
from heapq import heappop, heapify, heappush

class HighLevelStack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0
    
    def push(self, item: HighLevelNode):
        self.stack.append(item)
    
    def get_top_node(self):
        node = self.stack.pop()
        return node
    
    @property
    def size(self):
        return len(self.stack)
```
